# Plot of the points generated

This notebook shows illustrations of points generated with a non-fixed seed using GPU, OMP and CPU (seq) with the following distributions: normal, uniform and sphere/displaced sphere (p=0.1).

## Normal Distribution

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

# Function to read .off file and extract points
def read_off(file):
    if 'OFF' != file.readline().strip():
        raise 'Not a valid OFF header'
    n_verts, n_faces, n_edges = tuple([int(s) for s in file.readline().strip().split(' ')])
    vertices = [
        [float(s) for s in file.readline().strip().split(' ')]
        for i_vert in range(n_verts)
    ]
    return vertices

# Reading the .off file
with open('points_normal_gpu.off', 'r') as file:
    points = read_off(file)

# Extracting x, y, and z coordinates
x, y, z = zip(*[(pt[0], pt[1], pt[2]) for pt in points])

# Plotting in 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z)
ax.set_title('Parallel GPU generation using a Normal Distribution')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```
